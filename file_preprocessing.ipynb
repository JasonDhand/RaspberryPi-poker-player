```python
import os
import shutil

# Define the paths
source_folder = "cards_dataset/train"

# Iterate through the source folders
for card in ["ace", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten", "jack", "queen", "king"]:
    destination_folder = f"cards_dataset_by_rank/train/{card}"  # Define destination outside the inner loop
    
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
        print(f"Created folder: {destination_folder}")

    for suit in ["spades", "clubs", "diamonds", "hearts"]:
        # print("hello")
        suit_folder = os.path.join(source_folder, f"{card} of {suit}")
        print(suit_folder)
        # Iterate through the files in each suit folder
        for filename in os.listdir(suit_folder):
            # print("hello")
            source_file = os.path.join(suit_folder, filename)
            # If a file with the same name exists in the destination folder, rename it
            dest_file = os.path.join(destination_folder, filename)
            if os.path.exists(dest_file):
                name, ext = os.path.splitext(filename)
                counter = 1
                while os.path.exists(os.path.join(destination_folder, f"{name}_{counter}{ext}")):
                    counter += 1
                dest_file = os.path.join(destination_folder, f"{name}_{counter}{ext}")
            # Move the file to the destination folder
            shutil.move(source_file, dest_file)
            print(f"Moved {filename} to {destination_folder}")

print("Files moved successfully!")
```
