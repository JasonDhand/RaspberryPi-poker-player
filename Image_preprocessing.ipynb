```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import tensorflow as tf
```

```python
def noise_reduction(image):
    # Apply Gaussian blur with a kernel size of (5, 5)
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
    return blurred_image
```

```python
def preprocess_images(input_folder, output_folder):
    # if not os.path.exists(output_folder):
    #     os.makedirs(output_folder)

    # Iterate through the images in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            # Read the image
            image_path = os.path.join(input_folder, filename)
            image = cv2.imread(image_path)

            
            #gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            gaus_image = noise_reduction(image)
            # Resize the image
            resized_image = cv2.resize(gaus_image, (224, 224))

            # Save the preprocessed image to the output folder
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, resized_image)
            print(f"Preprocessed {filename} and saved to {output_path}")
```

```python
input_folder = "playing-cards\\img"
output_folder = "preprocessed_images"
preprocess_images(input_folder, output_folder)
```
